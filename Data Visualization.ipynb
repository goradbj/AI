### Matplotlib

```python
from matplotlib import pyplot as plt
import numpy as np
```

#### Basics of plotting

```python
# Line Plot
x=[1,2,3]
y=[2,4,1]

plt.plot(x,y)

plt.title("Simple Line plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid()
plt.show()
```

```python
## Plot sine wave using numpy and matplotlib

x=np.arange(0,10,0.1)
y=np.sin(x)

print(x)

print(y)
```

```python
y.max()
```

```python
y.min()
```

```python
# Sine Wave
plt.plot(x,y)
plt.title("Sine wave")
plt.xlabel("X Values")
plt.ylabel("Sin(X) values")
plt.show()
```

```python
# Cos Wave
x=np.arange(0,10,0.1)
y=np.cos(x)
plt.plot(x,y)
plt.title("Cos wave")
plt.xlabel("X Values")
plt.ylabel("Cos(X) values")
plt.show()
```

```python
# tan Wave
x=np.arange(0,10,0.1)
y=np.tan(x)
plt.plot(x,y)
plt.title("Tan wave")
plt.xlabel("X Values")
plt.ylabel("Tan(X) values")
plt.show()
```

```python
# Cot Wave
x=np.arange(0,10,0.1)
y=1/np.tan(x)
plt.plot(x,y)
plt.title("Cot wave")
plt.xlabel("X Values")
plt.ylabel("Cot(X) values")
plt.show()
```

```python
# Sec Wave
x=np.arange(0,10,0.1)
y=1/np.cos(x)
plt.plot(x,y)
plt.title("Sec wave")
plt.xlabel("X Values")
plt.ylabel("Sec(X) values")
plt.show()
```

```python
# Cosec Wave
x=np.arange(0,10,0.1)
y=1/np.sin(x)
plt.plot(x,y)
plt.grid()
plt.title("Cosec wave")
plt.xlabel("X Values")
plt.ylabel("Cosec(X) values")
plt.show()
```

### Text and Annotations

### 3D Plotting

```python
from mpl_toolkits import mplot3d
import numpy as np
from matplotlib import pyplot as plt
```

```python
# how to plot empty 3d graph

%matplotlib notebook
fig=plt.figure()
ax=plt.axes(projection="3d")
```

```python
# how to generate 3D line plot
x=np.linspace(0,15,100)
y=np.sin(x)
z=np.cos(x)
print(x)
print()
print(y)
print()
print(z)
%matplotlib notebook
fig=plt.figure()
ax=plt.axes(projection="3d")
ax.plot3D(x,y,z)
```

```python
# how to generate 3D scatter plot
x=np.linspace(0,15,100)
y=np.sin(x)
z=np.cos(x)

print(x)
print()
print(y)
print()
print(z)

%matplotlib notebook
fig=plt.figure()
ax=plt.axes(projection="3d")
ax.scatter3D(x,y,z)
```
