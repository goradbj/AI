### Import the required Libraries

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

### Load the Dataset

```python
iris=load_iris()
```

```python
X=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
X
```

```python
y=pd.Categorical.from_codes(iris.target,iris.target_names)
```

```python
y
```

### Explore the dataset

```python
X.head()
```

```python
X.shape
```

```python
y.shape
```

```python
X.info()
```

```python
y=pd.get_dummies(y)
```

```python
y
```

### Divide the data into training and testing

```python
X_tr,X_te,y_tr,y_te=train_test_split(X,y,test_size=0.3)
```

```python
print(X_tr.shape)
print(X_te.shape)
print(y_tr.shape)
print(y_te.shape)
```

### Build model now

```python
dtmodel1=DecisionTreeClassifier()
```

```python
dtmodel1.fit(X_tr,y_tr)
```

```python
dtmodel2=DecisionTreeClassifier(criterion='entropy')
```

```python
dtmodel2.fit(X_tr,y_tr)
```

### do the Prediction

```python
y_pred=dtmodel1.predict(X_te)
```

```python
y_pred
```

```python
y_te
```

### Evaluate the model

```python
species=np.array(y_te).argmax(axis=1)

pred=np.array(y_pred).argmax(axis=1)

confusion_matrix(species,pred)
```
